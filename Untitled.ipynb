```python
# find duplicates
# select duplicate rows except first occurrence based on all columns
def duplicates(df):
    duplicate_rows_df = df[df.duplicated()]
    return duplicate_rows_df
```

```python
duplicates(df)
```

```python
df.drop_duplicates(inplace = True)
```

```python
# confirm duplicates removed
duplicates(df)
```
